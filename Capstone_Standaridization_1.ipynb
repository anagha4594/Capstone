```python
# importing the libraries
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
import warnings
```

```python
# importing the G20 data with null value filtered from columns 1960-61
data=pd.read_excel('G20_filtered.xlsx',sheet_name='Sheet1',encoding='cp1252')
data.head()                
```

```python
# shape of original dataset
data.shape
```

```python
# number of Countries in dataset
data['Country Name'].nunique()
```

```python
# Data summary Stats
data.describe(include='all')
```

```python
# Transposing the year and indicators
new1=pd.DataFrame()
g20=data['Country Name'].unique()

for i in g20:

    A=data[data['Country Name']==i].drop(['Country Name','Country Code','Category','Indicator Code'],axis=1)
    #A1['Country Name']=i
    A1=A.drop(['Indicator Name'],axis=1).T
    A1.columns=A['Indicator Name']
    A1['Country Name']=i
    new1=pd.concat([new1,A1],axis=0)
#A1.drop('Country Name',axis=1)
new1.head()
```

```python
# Shape of New Transposed G20 dataset
new1.shape
```

```python
# Writing the dataset back to a csv file
new1.to_csv('Edited.csv',index=True)
```

# Standardizing the Values using StandardScaler()

```python
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
country=data[['Year','Country Name']]
col=data.columns
col=col.drop(['Year','Country Name'])
data_std=sc.fit_transform(data.drop(['Year','Country Name'],axis=1))

# Copying the standardized array into a Dataframe
data_std=pd.DataFrame(data_std)
data_std.columns=col
```

```python
# Standardized Dataset
data_std=pd.concat([country,data_std],axis=1)
data_std.head()
```

```python
# Writing back the file to a CSV file
data_std.to_csv('Edited_std.csv',index=False)
```
